```python
import os;
import io;
import pandas as pd; 
import gzip

files = os.listdir()

for files in files:
    if (files.endswith('.txt')):
        clinical = pd.read_csv(files, sep = '\t')
        print(clinical.head(10))
```

```python
#number of rows
print(clinical.shape[0])

#number of columms
print(clinical.shape[1])

#number of unique patients
clinical[['patient']].nunique()
```

```python
clinical_short = clinical[['patient' , 'year_of_birth' , 'gender' , 'race' , 'definition' , 'tumor_stage'  , 'age_at_diagnosis' , 'vital_status' , 'days_to_death', 'days_to_last_follow_up']]
```

```python
#checking my sanity... it's there!
clinical_short.head()
```

```python
#convert age_at_diagnosis from days to years, round to nearest hundredth and overwrite original values -- don't really know why this needs to be added, but it doesn't run without it. 
pd.options.mode.chained_assignment = None  # default='warn'


age_messing = clinical_short['age_at_diagnosis']/365
clinical_short['age_at_diagnosis'] = round(age_messing , 2)
```

```python
dead = clinical_short['days_to_death']/12
clinical_short['days_to_death'] = round(dead , 2)
```

```python
doctorz_visit = clinical_short['days_to_last_follow_up']/12
clinical_short['days_to_last_follow_up'] = round(doctorz_visit , 2)
```

```python
clinical_short.rename(columns = {'days_to_death':'months_to_death' , "days_to_last_follow_up": "months_to_last_follow_up"} , inplace = True)
```

```python
#let's do another sanity check
clinical_short
```

```python
#remove healthy patients from clinical_short
clinical_short_for_export = clinical_short[clinical_short.definition != 'Solid Tissue Normal']
```

```python
#Another quick sanity check.
clinical_short_for_export['definition'].unique()
```

```python
clinical_short_for_export.to_csv('clinical_short.csv' , sep = '\t')
```
